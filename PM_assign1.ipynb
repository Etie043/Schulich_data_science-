```python
# Importing the necessary libraries
import pandas as pd
import statsmodels.api as sm
import numpy as np
from statsmodels.tsa.stattools import acf
```

```python
# Loading the dataset
data = pd.read_csv('C:/Users/etiem/Downloads/dataset_lm.csv')
```

```python
# Defining the dependent and independent variables
X = data.drop(columns=['Dependent Var'])  # Explanatory variables
y = data['Dependent Var']  # Dependent variable

# Adding a constant to the independent variables
X = sm.add_constant(X)

# Running the OLS model
model = sm.OLS(y, X).fit()

# Outputting the summary of the model
model_summary = model.summary()
print(model_summary)
```

Part B

```python
# Step 1: Calculate the residuals (errors) from the OLS model
residuals = model.resid

# Step 2: Calculate the standard deviation of the residuals
residuals_std = np.std(residuals)
print(f"Standard Deviation of Residuals: {residuals_std}")

# Step 3: Calculate the autocorrelation values for the first three lags
autocorrelation_values = acf(residuals, nlags=3)
print(f"Autocorrelation Values for Lags 1-3: {autocorrelation_values}")
```

```python
# Using autocorrelation at lag 1 (small value, might not significantly impact the model)
rho = autocorrelation_values[1]  # First lag autocorrelation (very small)

# Create a basic weights matrix for the GLS model, using rho for lag 1
weights = np.diag([1] * len(residuals))
for i in range(1, len(residuals)):
    weights[i, i-1] = -rho  # Applying lag 1 autocorrelation (though very small)

# Running the GLS model
gls_model = sm.GLS(y, X, sigma=weights).fit()

# Output the summary of the GLS model
gls_model_summary = gls_model.summary()
print(gls_model_summary)
```
